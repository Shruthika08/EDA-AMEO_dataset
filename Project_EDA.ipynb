# AMEO Dataset analysis

**Introduction:**

Performing Exploratory Data Analysis (EDA) on the given dataset by considering **Salary** as a target variable.

**Dataset description:**

The dataset was released by Aspiring Minds from the Aspiring Mind Employment Outcome 2015 (AMEO). The study is primarily limited  only to students with engineering disciplines. The dataset contains the employment outcomes of engineering graduates as dependent variables (Salary, Job Titles, and Job Locations) along with the standardized scores from three different areas â€“ cognitive skills, technical skills and personality skills. The dataset also contains demographic features. The dataset  contains  around  40 independent variables and 4000 data points. The independent variables are both continuous and categorical in nature. The dataset contains a unique identifier for each candidate.

**Objective:**

1. Offer detailed explanation about the dataset.
2. Extract the relationship between different variables and the target variable i.e salary.
3. Find the outliers and treat them accordingly.
4. Find any patterns in the data using the univariate analysis.

```python
def reducing_cat(df, col_name):
    for cat in df[col_name].unique():
        # if col_name == 'Designation' and cat in exception:
        #     continue
        cat_count = df[df[col_name] == cat][col_name].value_counts()[0]
        min_count = df[col_name].value_counts()[:10].min()
        if cat_count < min_count:
            df.loc[df[col_name] == cat, col_name] = 'other'

def print_top_categories(df, categorical_cols):
    for col in categorical_cols:
        reducing_cat(df, col)
        print('Top 10 categories in:', col)
        print('')
        print(df[col].value_counts())
        print('')
        print('-'*100)

# Example usage
print_top_categories(df, categorical_cols)
```

# Univariate analysis

Numerical data:

```python
def univariate_analysis_multiple_columns(data, column_list):
    for column_name in column_list:
        # Descriptive statistics
        pd.options.display.float_format = '{:.2f}'.format
        print(f"Descriptive statistics for {column_name}:")
        print("=======================================")
        print(data[column_name].describe())
        print("\n")

        # Observations
        print(f"Observations for {column_name}:")
        print("=======================================")
        print("Mean:", data[column_name].mean())
        print("Median:", data[column_name].median())
        print("Standard Deviation:", data[column_name].std())
        print("\n")

        import seaborn as sns
        import matplotlib.pyplot as plt

        # Plot PDF
        sns.displot(data[column_name], kind='kde', rug=True, color='blue')
        plt.title(f'Probability Density Function (PDF) of {column_name}', color='blue')
        plt.show()

        # Plot histogram
        sns.histplot(data[column_name], kde=True, color='green')
        plt.title(f'Histogram of {column_name}', color='green')
        plt.show()

        # Plot boxplot
        sns.boxplot(x=data[column_name], color='orange')
        plt.title(f'Boxplot of {column_name}', color='orange')
        plt.show()


        # Define function to calculate ECDF
        def ecdf(data):
            x = np.sort(data)
            y = np.arange(1, len(x) + 1) / len(x)
            return x, y

        # Generate random data (replace this with your actual data)
        df2 = data[column_name]
        sample = np.random.normal(df2.mean(), df2.std(), size=len(df2))

        x_, y_ = ecdf(df2)
        x_sample, y_sample = ecdf(sample)

        # Plot ECDF
        plt.figure(figsize=(5, 4))
        plt.plot(x_, y_, linestyle='-', marker='.', color='black', alpha=0.7, label= column_name)
        plt.plot(x_sample, y_sample, linestyle='-', marker='.', color='blue', alpha=0.7, label='Normal Distribution')
        plt.xlabel(column_name)
        plt.ylabel('ECDF')
        plt.legend()
        plt.tight_layout()
        plt.show()
```

```python
num_cols = ['Salary', '10percentage','12percentage', 'collegeGPA', 'ComputerProgramming', 'English', 'Logical', 'Quant', 'age', 'tenure']

univariate_analysis_multiple_columns(df, num_cols)
```

# Bivariate Analysis:

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

def bivariate_analysis(data, column_list):
    for column_name in column_list:
        # Plot scatterplot for numerical-numerical
        sns.scatterplot(x=column_name, y='Salary', data=data)
        plt.title(f'Scatterplot of {column_name} vs Salary')
        plt.xlabel(column_name)
        plt.ylabel('Salary')
        plt.show()

        # Calculate correlation coefficient
        correlation_coefficient = data['Salary'].corr(data[column_name])
        print(f"Correlation coefficient between Salary and {column_name}: {correlation_coefficient}")

        # Calculate Spearman correlation coefficient
        spearman_corr_coefficient = data['Salary'].corr(data[column_name], method='spearman')
        print(f"Spearman correlation coefficient between Salary and {column_name}: {spearman_corr_coefficient}")

    print("\n")


```

```python
col_list = ['10percentage','12percentage', 'collegeGPA', 'age', 'tenure', 'English', 'Logical', 'Quant']

bivariate_analysis(df, col_list)
```

# Relationship between gender and specialisation




```python
obsr = pd.crosstab(df.Specialization,df.Gender)
obsr
```

```python
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(df['Gender'], df['Specialization'])
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("There is a significant relationship between gender and specialization.")
else:
    print("There is no significant relationship between gender and specialization.")
```

```python
from statistics import mean
salary = []
for i in df.index:
  if df.loc[i, 'Specialization'] == 'computer engineering':
    if df.loc[i, 'Designation'] == 'hardware engineer' or df.loc[i, 'Designation'] == 'associate engineer' or df.loc[i, 'Designation'] == 'programmer analyst' or df.loc[i, 'Designation'] == 'software engineer':
      salary.append(df.loc[i, 'Salary'])

print(mean(salary))
```
